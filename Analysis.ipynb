```python
from datascience import *
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = Table.read_table('Combined Weeks.csv')
data.show(11)
```

# Sleep vs Productivity

```python
data.scatter("Sleep (hrs)", "% Completed")
```

##### In this analysis, I created a scatter plot to examine the relationship between sleep duration and my productivity during study sessions. Sleep duration (in hours) is plotted on the x-axis, while productivity, measured as the percentage of homework completed, is plotted on the y-axis. Each point represents a single study session. The plot shows a very weak correlation between sleep and % work complete.

# Trend Over Time

```python
data = data.with_column("Session", np.arange(1, data.num_rows + 1))
data.show()
```

##### To make a line graph I create a new session column to track numerical data for each week

```python
data.plot("Session", "% Completed")
```

###### From the graph we can see a slow upward trend in my productivity (% of work complete). However, the line tracks both days where I studied at home and at school, resulting in the dip in productivity each 2nd session.

# Trend Over Time (Home and School Seperated)

```python
school = data.where("Location", "School")
home = data.where("Location", "Home")

school.plot("Session", "% Completed")
home.plot("Session", "% Completed")
```

##### In this analysis, I separated my data into two groups based on study location (school and home) and created line charts to observe productivity trends within each environment. Each chart shows the percentage of homework completed across study sessions. The top graph represents school sessions, where productivity remains consistently high, generally staying between 85% and 92%. Although there are small fluctuations, the overall performance is stable and consistently strong. The bottom graph represents home sessions, where productivity is noticeably lower but shows a gradual upward trend over time. This suggests that while I may be improving slightly when studying at home, my performance still does not reach the same level as when studying at school.

# Location vs Productivity

```python
data.group("Location", np.mean).select("Location", "% Completed mean").barh("Location")
```

##### In this analysis, I grouped my data by study location and calculated the average percentage of homework completed in each location. I then used a bar chart to compare my average productivity when studying at home versus at school. The chart shows that my average productivity was higher when studying at school than when studying at home. This supports my hypothesis that study environment affects my ability to complete work efficiently. Even though I kept study duration and other conditions relatively consistent, location still appears to have had a noticeable impact on my performance.
